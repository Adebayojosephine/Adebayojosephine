```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", rc={'axes.grid': False})
# Scatter Plot for MSSubClass against SalePrice
plt.figure(figsize=(10, 5))
sns.scatterplot(x='MSSubClass', y='SalePrice', data=data)
plt.title('MSSubClass vs Sale Price')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='LotFrontage', y='SalePrice', data=data)
plt.title('LotFrontage and Sale Price')
plt.show()
```

```python
plt.figure(figsize=(14, 5))
sns.boxplot(x='OverallQual', y='SalePrice', hue='MSZoning', data=data)
plt.title('Overall Quality and Zoning vs Sale Price')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.violinplot(x='MSZoning', y='SalePrice', data=data)
plt.title('Zoning vs Sale Price')
plt.show()
```
